```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\_PROJECT_\Exploratory Data Analysis\globalterrorism.csv', encoding='ISO-8859-1')
df.head()
```

```python
categorical_columns = ['country', 'region', 'attacktype1_txt']

for col in categorical_columns:
    unique_values = df[col].unique()
    print(f"\nUnique values in {col}:")
    print(unique_values)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.countplot(data=df, x='iyear')
plt.xticks(rotation=-270)
plt.title('Terrorist Attacks by Year')
plt.plot()
```

```python

# Distribution of attack types
# plt.subplot(2, 2, 2)
sns.countplot(data=df, y='attacktype1_txt', order=df['attacktype1_txt'].value_counts().index)
plt.title('Distribution of Attack Types')
plt.show()
```

```python
sns.countplot(data=df, y='region_txt', order=df['region_txt'].value_counts().index)
plt.title('Number of Attacks in Different Regions')
plt.show()
```

```python
sns.boxplot(data=df, x='attacktype1_txt', y='nkill')
plt.xticks(rotation=45)
plt.title('Number of Casualties by Attack Type')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='imonth')
plt.title('Distribution of Terrorist Attacks by Month')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='success')
plt.title('Distribution of Attack Success/Failure')
plt.xlabel('Attack Success/Failure')
plt.ylabel('Number of Attacks')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='nkill', bins=30, kde=True)
plt.title('Histogram of Casualties')
plt.xlabel('Number of Casualties')
plt.ylabel('Number of Attacks')
```

```python
plt.figure(figsize=(10, 6))
top_industries = df['targtype1_txt'].value_counts().head(10)
sns.barplot(x=top_industries, y=top_industries.index)
plt.title('Top 10 Targeted Industries')
plt.xlabel('Number of Attacks')
plt.ylabel('Industry')
```
